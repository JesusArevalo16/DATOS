```python
automoviles = df[['Placa', 'Modelo']].drop_duplicates().reset_index(drop=True)
automoviles['auomoviles_id'] = automoviles.index + 1

# Normalización: Crear tabla de clientes
clientes = df[['Estacionamiento']].drop_duplicates().reset_index(drop=True)
clientes.rename(columns={'Estacionamiento': 'Estacionamiento_ID'}, inplace=True) # Renaming for clarity as a FK
clientes['cliente_nombre'] = 'Cliente ' + (clientes['Estacionamiento_ID']).astype(str)

# Normalización: Crear tabla de estacionamiento (details about parking events)
estacionamiento_registros = df[['Id_registro', 'Estacionamiento']].drop_duplicates().reset_index(drop=True)


estacionamiento_detallado = estacionamiento_registros.merge(
    clientes,
    left_on='Estacionamiento',
    right_on='Estacionamiento_ID',
    how='left'
).drop(columns=['Estacionamiento_ID']) # Drop the redundant ID column from the merge

print("\nTabla Automóviles (Normalizada):\n", automoviles)
print("\nTabla Clientes (Normalizada):\n", clientes) # Print the new 'clientes' table
print("\nTabla Estacionamiento Registros (Normalizada):\n", estacionamiento_detallado) # Print the detailed estacionamiento table
```

```python
# Unir (Desnormalizar) las tablas.
temp_df_denormalizado = pd.merge(
    estacionamiento_detallado,
    df[['Id_registro', 'Hora_entrada', 'Hora_salida', 'Placa', 'Modelo']],
    on='Id_registro',
    how='left'
)

df_desnormalizado = pd.merge(
    temp_df_denormalizado,
    automoviles[['Placa', 'Modelo', 'auomoviles_id']],
    on=['Placa', 'Modelo'],
    how='left'
)

print("\nDatos Desnormalizados (Unidos):\n", df_desnormalizado)
```
